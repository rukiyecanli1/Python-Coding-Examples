```python
# n.a, -5 ve cinsiyetteki Efe değeri de NaN olarak kabul edilmelidir.
#Bunun için;

df=pd.read_table("9_data4.txt",sep=",",
                na_values=["n.a",-5,"Efe"]) 
df
```

```python
#Yukarıdaki işlemde Efe'yi eksik veri olarak aldık fakat isism de NaN oldu bunu istemiyoruz.
#Bunun için eksik verileri sözlük yapısında yazabiliriz.

eksik_veri={"puan":[-5],"cinsiyet":["n.a","Efe"]}
df=pd.read_table("9_data4.txt",sep=",",
                na_values=eksik_veri) 
df
```
